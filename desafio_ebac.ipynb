Preparando biblioteca

```python
import pandas as pd
```

Preparando dataset

```python
data_file = 'dataset.xlsx'
df = pd.read_excel(data_file)
df.info()
```

8 variáveis numéricas: Unnamed (índice), preço, unidades vendidas,cidade_alcool,cidade_gasolina,estrada_alcool,estrada_gasolina,custo_revisao

Resumo estatístico das variáveis

```python
print(df.describe())
```

5 carros que mais consomem gasolina na cidade

```python
df_ordenado = df.sort_values(by='cidade_gasolina', ascending=True)
top_5_consumo = df_ordenado.head(5)
```

```python
print(top_5_consumo[['modelo', 'cidade_gasolina']])
```

Média de consumo de álcool tanto na cidade quanto na estrada

```python
media_cidade_alcool = df['cidade_alcool'].mean()
media_estrada_alcool = df['estrada_alcool'].mean()
```

```python
print(f"Média na Cidade: {media_cidade_alcool:.2f} km/l")
print(f"Média na Estrada: {media_estrada_alcool:.2f} km/l")
```

Nome da montadora para cada carro

```python
df['montadora'] = df['modelo'].str.split().str[0]
print(df[['modelo', 'montadora', 'preco']].head())
```

5 montadoras com a maior média de preço de carros

```python
media_preco_por_montadora = df.groupby('montadora')['preco'].mean()
montadoras_ordenadas = media_preco_por_montadora.sort_values(ascending=False)
top_5_maiores_medias = montadoras_ordenadas.head(5)
print(top_5_maiores_medias)
```

10 carros com menor custo de manutenção

```python
df_ordenado_custo = df.sort_values(by='custo_revisao', ascending=True)
menor_custo_manutencao = df_ordenado_custo.head(10)
print(menor_custo_manutencao[['modelo', 'custo_revisao']])
```

Carro mais vendido para cada montadora

```python
idx_mais_vendido = df.groupby('montadora')['unidades_vendidas'].idxmax()
carros_mais_vendidos = df.loc[idx_mais_vendido]
resultado_final = carros_mais_vendidos[['montadora', 'modelo', 'unidades_vendidas']].sort_values('unidades_vendidas', ascending=False)
print(resultado_final)
```

Super desafio

```python
# Custo da Gasolina
preco_gasolina_litro = 6.03  # Valor médio para o RJ
consumo_litros_mes = 150
total_meses = 24
custo_total_gasolina_2_anos = preco_gasolina_litro * consumo_litros_mes * total_meses

# Custo do IPVA
taxa_ipva_anual = 0.04
total_anos_ipva = 2

# Custo da Revisão
revisoes_por_ano = 1
total_anos_revisao = 2
```

Costo de IPVA em 2 anos

```python
df['custo_ipva_2_anos'] = df['preco'] * taxa_ipva_anual * total_anos_ipva
```

Custo de Revisão em 2 anos

```python
df['custo_revisoes_2_anos'] = df['custo_revisao'] * revisoes_por_ano * total_anos_revisao
```

Coluna final somando preço inicial + todos os custos de 2 anos

```python
df['custo_total_em_2_anos'] = (
    df['preco'] +
    df['custo_ipva_2_anos'] +
    df['custo_revisoes_2_anos'] +
    custo_total_gasolina_2_anos
)
```

```python
df_resultado = df.sort_values(by='custo_total_em_2_anos', ascending=True)
```

```python
print(f"Custo fixo de gasolina para 2 anos (150L/mês a R${preco_gasolina_litro}/L): R$ {custo_total_gasolina_2_anos:,.2f}\n")
print("--- Custo Total de Propriedade em 2 Anos (Preço + Despesas) ---")
print(df_resultado[['modelo', 'preco', 'custo_total_em_2_anos']])
```
